```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
exercise = pd.read_csv('exercise.csv')
calories = pd.read_csv('calories.csv')
df = pd.merge(exercise, calories, on='User_ID' )
```

```python
sns.pairplot(df, hue='Gender')
```

```python
distributions_columns = ['Age', 'Height', 'Weight', 'Duration', 'Calories']
for distribution in distributions_columns:
    plt.figure(figsize=(12,6))
    plt.hist(x=df[distribution])
    plt.title(f"Distribution of Data Points of {distribution}")
    
```

```python
df['Gender'].value_counts()
```

```python
sns.boxplot(x=df['Gender'], y=df['Calories'])
```

```python
plt.figure(figsize=(16,6))
plt.yticks(rotation=90)
sns.boxplot(x=df['Age'], y=df['Calories'])
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df['Weight'], y=df['Calories'], hue=df['Gender'])
```

```python
sns.jointplot(x=df['Duration'], y=df['Calories'],kind='scatter')
```

```python
sns.jointplot(x=df['Heart_Rate'], y=df['Calories'],kind='hex')
```
